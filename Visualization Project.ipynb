## Matrix Factorization

Original implementation $\lambda/2(u^2 + v^2)$ + 1/2 sum (Y_ij - uTv)^2

```python
def grad_U(Ui, Yij, Vj, reg, eta):
    """
    Takes as input Ui (the ith row of U), a training point Yij, the column
    vector Vj (jth column of V^T), reg (the regularization parameter lambda),
    and eta (the learning rate).

    Returns the gradient of the regularized loss function with
    respect to Ui multiplied by eta.
    """
    return (reg * Ui + (Yij - np.dot(Ui.T, Vj)) * -Vj) * eta

def grad_V(Vj, Yij, Ui, reg, eta):
    """
    Takes as input the column vector Vj (jth column of V^T), a training point Yij,
    Ui (the ith row of U), reg (the regularization parameter lambda),
    and eta (the learning rate).

    Returns the gradient of the regularized loss function with
    respect to Vj multiplied by eta.
    """
    return (reg * Vj + (Yij - np.dot(Ui.T, Vj)) * -Ui.T) * eta

def get_err(U, V, Y, reg=0.0):
    """
    Takes as input a matrix Y of triples (i, j, Y_ij) where i is the index of a user,
    j is the index of a movie, and Y_ij is user i's rating of movie j and
    user/movie matrices U and V.

    Returns the mean regularized squared-error of predictions made by
    estimating Y_{ij} as the dot product of the ith row of U and the jth column of V^T.
    """
    U_f = np.linalg.norm(U, ord='fro')
    V_f = np.linalg.norm(V, ord='fro')
    
    err = reg / 2 * (U_f**2 + V_f**2)
    sq_term_tot = 0
    
    for row in range(len(Y)):
        i, j, Y_ij = Y[row]
        i -= 1
        j -= 1
        estimate = np.dot(U[i], V.T[:, j])
        sq_term = (Y_ij - estimate) ** 2
        sq_term_tot += sq_term * 1/2
        
    return (err + sq_term_tot) / len(Y)

def train_model(M, N, K, eta, reg, Y, eps=0.0001, max_epochs=300):
    """
    Given a training data matrix Y containing rows (i, j, Y_ij)
    where Y_ij is user i's rating on movie j, learns an
    M x K matrix U and N x K matrix V such that rating Y_ij is approximated
    by (UV^T)_ij.

    Uses a learning rate of <eta> and regularization of <reg>. Stops after
    <max_epochs> epochs, or once the magnitude of the decrease in regularized
    MSE between epochs is smaller than a fraction <eps> of the decrease in
    MSE after the first epoch.

    Returns a tuple (U, V, err) consisting of U, V, and the unregularized MSE
    of the model.
    """
    U = np.random.random((M, K)) - 0.5
    V = np.random.random((N, K)) - 0.5
    
    first_epoch_mse = 0
    prev_epoch_mse = get_err(U, V, Y, reg)
    mse_change_after_first_epoch = 0
    
    for epoch in range(max_epochs):
        if (epoch % 5 == 0):
            print(epoch)
        for row in range(len(Y)):
            i, j, Yij = Y[row]
            i -= 1
            j -= 1
            U[i] = U[i] - grad_U(U[i], Yij, V[j], reg, eta)
            V[j] = V[j] - grad_V(V[j], Yij, U[i], reg, eta)
        after_epoch_mse = get_err(U, V, Y, reg)
        if epoch == 0: # 1st epoch
            first_epoch_mse = after_epoch_mse
            mse_change_after_first_epoch = prev_epoch_mse - first_epoch_mse
            prev_epoch_mse = first_epoch_mse
        else:
            mse_change = prev_epoch_mse - after_epoch_mse
            prev_epoch_mse = after_epoch_mse
            if (mse_change < eps * mse_change_after_first_epoch):
                break
    
    return U, V, get_err(U, V, Y, reg)
```

## Bias Term (a and b) Implementation

```python
def grad_U(Ui, Yij, Vj, ai, bj, reg, eta):
    """
    Takes as input Ui (the ith row of U), a training point Yij, the column
    vector Vj (jth column of V^T), reg (the regularization parameter lambda),
    and eta (the learning rate).

    Returns the gradient of the regularized loss function with
    respect to Ui multiplied by eta.
    """
    return (reg * Ui + (Yij - np.dot(Ui.T, Vj) - ai - bj) * -Vj)* eta

def grad_V(Vj, Yij, Ui, ai, bj, reg, eta):
    """
    Takes as input the column vector Vj (jth column of V^T), a training point Yij,
    Ui (the ith row of U), reg (the regularization parameter lambda),
    and eta (the learning rate).

    Returns the gradient of the regularized loss function with
    respect to Vj multiplied by eta.
    """
    return (reg * Vj + (Yij - np.dot(Ui.T, Vj) - ai - bj) * -Ui.T) * eta

def grad_a_or_b(ai, bj, Ui, Yij, Vj, reg, eta): # Works for both.
    return (Yij - np.dot(Ui.T, Vj) - ai - bj) * -1 * eta

def get_err(U, V, Y, a, b, reg=0.0):
    """
    Takes as input a matrix Y of triples (i, j, Y_ij) where i is the index of a user,
    j is the index of a movie, and Y_ij is user i's rating of movie j and
    user/movie matrices U and V, and vectors of bias/offset terms a and b. a has length
    number of users, and b has length number of movies.

    Returns the mean regularized squared-error of predictions made by
    estimating Y_{ij} as the dot product of the ith row of U and the jth column of V^T.
    """
    U_f = np.linalg.norm(U, ord='fro')
    V_f = np.linalg.norm(V, ord='fro')
    
    err = reg / 2 * (U_f**2 + V_f**2)
    sq_term_tot = 0
    
    for row in range(len(Y)):
        i, j, Y_ij = Y[row]
        i -= 1
        j -= 1
        estimate = np.dot(U[i], V.T[:, j])
        sq_term = (Y_ij - estimate - a[i] - b[j]) ** 2
        sq_term_tot += sq_term * 1/2
        
    return (err + sq_term_tot) / len(Y)

def train_model(M, N, K, eta, reg, Y, eps=0.0001, max_epochs=300):
    """
    Given a training data matrix Y containing rows (i, j, Y_ij)
    where Y_ij is user i's rating on movie j, learns an
    M x K matrix U and N x K matrix V such that rating Y_ij is approximated
    by (UV^T)_ij.

    Uses a learning rate of <eta> and regularization of <reg>. Stops after
    <max_epochs> epochs, or once the magnitude of the decrease in regularized
    MSE between epochs is smaller than a fraction <eps> of the decrease in
    MSE after the first epoch.

    Returns a tuple (U, V, err) consisting of U, V, and the unregularized MSE
    of the model.
    """
    U = np.random.random((M, K)) - 0.5
    V = np.random.random((N, K)) - 0.5
    a = np.random.random((M,)) - 0.5
    b = np.random.random((N,)) - 0.5
    
    first_epoch_mse = 0
    prev_epoch_mse = get_err(U, V, Y, a, b, reg)
    mse_change_after_first_epoch = 0
    
    for epoch in range(max_epochs):
        if (epoch % 5 == 0):
            print(epoch)
        for row in range(len(Y)):
            i, j, Yij = Y[row]
            i -= 1
            j -= 1
            U[i] = U[i] - grad_U(U[i], Yij, V[j], a[i], b[j], reg, eta)
            V[j] = V[j] - grad_V(V[j], Yij, U[i], a[i], b[j], reg, eta)
            a[i] = a[i] - grad_a_or_b(a[i], b[j], U[i], Yij, V[j], reg, eta)
            b[i] = b[i] - grad_a_or_b(a[i], b[j], U[i], Yij, V[j], reg, eta)
        after_epoch_mse = get_err(U, V, Y, a, b, reg)
        if epoch == 0: # 1st epoch
            first_epoch_mse = after_epoch_mse
            mse_change_after_first_epoch = prev_epoch_mse - first_epoch_mse
            prev_epoch_mse = first_epoch_mse
        else:
            mse_change = prev_epoch_mse - after_epoch_mse
            prev_epoch_mse = after_epoch_mse
            if (mse_change < eps * mse_change_after_first_epoch):
                break
    
    return U, V, a, b, get_err(U, V, Y, a, b, reg)
```

## a and b both terms implementation

```python
def grad_U(Ui, Yij, Vj, ai, bj, reg, eta):
    """
    Takes as input Ui (the ith row of U), a training point Yij, the column
    vector Vj (jth column of V^T), reg (the regularization parameter lambda),
    and eta (the learning rate).

    Returns the gradient of the regularized loss function with
    respect to Ui multiplied by eta.
    """
    return (reg * Ui + (Yij - np.dot(Ui.T, Vj) - ai - bj) * -Vj)* eta

def grad_V(Vj, Yij, Ui, ai, bj, reg, eta):
    """
    Takes as input the column vector Vj (jth column of V^T), a training point Yij,
    Ui (the ith row of U), reg (the regularization parameter lambda),
    and eta (the learning rate).

    Returns the gradient of the regularized loss function with
    respect to Vj multiplied by eta.
    """
    return (reg * Vj + (Yij - np.dot(Ui.T, Vj) - ai - bj) * -Ui.T) * eta

def grad_a(ai, bj, Ui, Yij, Vj, reg, eta):
    return (reg * ai + (Yij - np.dot(Ui.T, Vj) - ai - bj) * -1) * eta

def get_err(U, V, Y, a, b, reg=0.0):
    """
    Takes as input a matrix Y of triples (i, j, Y_ij) where i is the index of a user,
    j is the index of a movie, and Y_ij is user i's rating of movie j and
    user/movie matrices U and V, and vectors of bias/offset terms a and b. a has length
    number of users, and b has length number of movies.

    Returns the mean regularized squared-error of predictions made by
    estimating Y_{ij} as the dot product of the ith row of U and the jth column of V^T.
    """
    U_f = np.linalg.norm(U, ord='fro')
    V_f = np.linalg.norm(V, ord='fro')
    a_f = np.linalg.norm(a, ord='fro')
    b_f = np.linalg.norm(b, ord='fro')
    
    err = reg / 2 * (U_f**2 + V_f**2 + a_f**2 + b_f**2)
    sq_term_tot = 0
    
    for row in range(len(Y)):
        i, j, Y_ij = Y[row]
        i -= 1
        j -= 1
        estimate = np.dot(U[i], V.T[:, j])
        sq_term = (Y_ij - estimate + a[i] + b[j]) ** 2
        sq_term_tot += sq_term * 1/2
        
    return (err + sq_term_tot) / len(Y)

def train_model(M, N, K, eta, reg, Y, eps=0.0001, max_epochs=300):
    """
    Given a training data matrix Y containing rows (i, j, Y_ij)
    where Y_ij is user i's rating on movie j, learns an
    M x K matrix U and N x K matrix V such that rating Y_ij is approximated
    by (UV^T)_ij.

    Uses a learning rate of <eta> and regularization of <reg>. Stops after
    <max_epochs> epochs, or once the magnitude of the decrease in regularized
    MSE between epochs is smaller than a fraction <eps> of the decrease in
    MSE after the first epoch.

    Returns a tuple (U, V, err) consisting of U, V, and the unregularized MSE
    of the model.
    """
    U = np.random.random((M, K)) - 0.5
    V = np.random.random((N, K)) - 0.5
    a = np.random.random((M,)) - 0.5
    b = np.random.random((N,)) - 0.5
    
    first_epoch_mse = 0
    prev_epoch_mse = get_err(U, V, Y, a, b, reg)
    mse_change_after_first_epoch = 0
    
    for epoch in range(max_epochs):
        if (epoch % 5 == 0):
            print(epoch)
        for row in range(len(Y)):
            i, j, Yij = Y[row]
            i -= 1
            j -= 1
            U[i] = U[i] - grad_U(U[i], Yij, V[j], a[i], reg, eta)
            V[j] = V[j] - grad_V(V[j], Yij, U[i], b[j], reg, eta)
            a[i] = a[i] - grad_a_or_b(a[i], b[j], U[i], Yij, V[j], reg, eta)
            b[i] = b[i] - grad_a_or_b(a[i], b[j], U[i], Yij, V[j], reg, eta)
        after_epoch_mse = get_err(U, V, Y, a, b, reg)
        if epoch == 0: # 1st epoch
            first_epoch_mse = after_epoch_mse
            mse_change_after_first_epoch = prev_epoch_mse - first_epoch_mse
            prev_epoch_mse = first_epoch_mse
        else:
            mse_change = prev_epoch_mse - after_epoch_mse
            prev_epoch_mse = after_epoch_mse
            if (mse_change < eps * mse_change_after_first_epoch):
                break
    
    return U, V, a, b, get_err(U, V, Y, a, b, reg)
```
